```python
def reset_seed(SEED = 0):   
    os.environ['PYTHONHASHSEED']=str(SEED)
    np.random.seed(SEED)
reset_seed(2022)
```

```python
def UndoSilentRegionRemoval(features):
    """
    Switch back the Raman vector index from the post-curation phase to the original, pre-curation
    We undo the effect of the silent region removal. 
    Needed to report the results in terms of the Raman wavelengths.
    """
    features = [x+104 for x in features]
    features = [x+467 if x>907 else x for x in features]
    return features

def ExpandOverSilentRegion(values):
    """
    Takes a vector of values over the reduced 1167 features and expands it to 1738 to have 
    the silent region removal undone, placing 0 values over the silent region. 

    # Remove silent regions, i.e., undo the code line below
    regions = [spectrum[104:907], spectrum[1374:]]    
    """
    # values[0:0] = [0] * 104
    # values[907:907] = [0] * (1374-907)
    # return values
    result = np.insert(values, 0, [0] * 104)
    result = np.insert(result, 907, [0] * (1374-907))
    return result


def RamanIndex2Wavelength(features):
    """
    Revert from the 0-1737 indices to the original Raman frequencies, for an easier experimental validation.
    """        
    index = [x+1 for x in features]
    return RamanFreqTable[RamanFreqTable["Index"].isin(index)]["Raman"].tolist()
```

```python
def discretize(data, multip = 1000):
    
    return (data * multip).astype(int)
```

```python
def feat_importance(data, labels, cutoff=0.2, verbose=False, CV_folds=CV5folds):
    """
    Extracts the most important features/wavelengths using ANOVA and CHI2.
    The cutoff parameter defines the threshold parameter for selecting the top features.
    It is used as a multiplicative factor off the normalized max. 
    
    We use a 5-fold cross-validation strategy, to help dampen the differences between the results on different subsets.
    The feature importance vectors from each fold are filtered: only those with p-value<0.05 are retained. 
    They are then added together, normalized, and then the top features selected.
    
    The data is a list of dictionaries, one dictionary for each sample. Each sample alread has its data flattened,
    with the shape being the number of spots in the sample times the Raman vector. 
    The labels are a list, one entry for each sample, of the same shape as the sample, except that instead of the 
    Raman vector, it has a single integer label. 
    """


    # Initialize the ANOVA f-score values between each feature/wavelength and class (tumor or non-tumor)
    f_scores = np.zeros(data[0][SAMPLE_DICT_DATA_KEY].shape[1])
    
    # Initialize the CHI2 statistics between each feature/wavelength and class (tumor or non-tumor)
    chi_scores = np.zeros(1167)
        
    # Loop through the CV folds
    for valid_fold in CV_folds:     
        
        # Get the train data for this CV fold
        train_X=np.empty( (0,data[0][SAMPLE_DICT_DATA_KEY].shape[1]) )
        train_y=np.empty( 0 )
        valid_names=[]
        for (sample,label) in zip(data,labels): 
            if sample[SAMPLE_DICT_NAME_KEY] in valid_fold:
                valid_names.append(sample[SAMPLE_DICT_NAME_KEY])
            else: 
                train_X = np.append(train_X, sample[SAMPLE_DICT_DATA_KEY], axis=0)
                train_y = np.append(train_y, label, axis=0)
        #if len(valid_names) != len(valid_fold):
        #    print("Error in the design of CV fold (data not found):", valid_fold)
        #    sys.exit(-1)     
            
        # Get the ANOVA F-values per wavelength for this CV fold. Keep only those with p-value less than 0.05.
        CV_f_scores, CV_p_f_scores = f_classif(train_X, train_y)
        CV_f_scores = np.where(CV_p_f_scores < 0.05, CV_f_scores, 0)    
        f_scores += CV_f_scores
        
        # Get the CHI2 values per wavelength for this CV fold. Keep only those with p-value less than 0.05.
        CV_chi_scores, CV_p_chi_scores = chi2(discretize(train_X), train_y)
        CV_chi_scores = np.where(CV_p_chi_scores < 0.05, CV_chi_scores, 0)    
        chi_scores += CV_chi_scores
        
    # Take the average values
    f_scores = f_scores/len(CV_folds)           
    chi_scores = chi_scores/len(CV_folds)

    # Normalize the feature importance vector for easier extraction
    f_scores = f_scores/max(f_scores)    
    chi_scores = chi_scores/max(chi_scores)
    
    if verbose: 
        plt.plot(RAMAN_FREQ_RANGE, ExpandOverSilentRegion(f_scores))
        plt.xlim([RAMAN_FREQ_RANGE[0], RAMAN_FREQ_RANGE[-1]])
        plt.axhline(y=cutoff, color='r', linestyle='-')
        plt.show()
        plt.plot(RAMAN_FREQ_RANGE, chi_scores)
        plt.axhline(y=cutoff, color='r', linestyle='-')
        plt.show()    
        
    # Get the indices of the important and relevant features: scores (average per fold) over the threshold
    f_scores_sort = np.argsort(f_scores)
    f_importance = f_scores_sort[f_scores[f_scores_sort] > cutoff]
    chi_scores_sort = np.argsort(chi_scores)
    c_importance = chi_scores_sort[chi_scores[chi_scores_sort] > cutoff]
    
    # Undo the silent region removal and sort to observe bands
    f_importance = sorted(UndoSilentRegionRemoval(f_importance))       
    c_importance = sorted(UndoSilentRegionRemoval(c_importance))  
    
    # Group the consecutive features into bands/intervals
    f_ranges =[]
    for k,g in groupby(enumerate(f_importance),lambda x:x[0]-x[1]):
        group = (map(itemgetter(1),g))
        group = list(map(int,group))
        f_ranges.append((group[0],group[-1]))
    
    c_ranges =[]
    for k,g in groupby(enumerate(c_importance),lambda x:x[0]-x[1]):
        group = (map(itemgetter(1),g))
        group = list(map(int,group))
        c_ranges.append((group[0],group[-1]))

    if verbose: 
        print("ANOVA:", len(f_importance), "features", f_ranges)
        print("Chi2:", len(c_importance), "features", c_ranges)

    return f_importance, f_scores, c_importance, chi_scores
```

```python
def model_feat_importance(ModelFolder, cutoff=0.2, verbose=False):
    """
    Extracts the most important features/wavelengths based on the pre-trained models in ModelFolder.
    Assumes that in ModelFolder we find the models corresponding to each fold in a cross-validation training.
    Assumes that this is all there is in that folder. 
    Assumes all models are already fit and it does not perform model fitting. 
    Adds the feature importances from each CV model, normalizes them with respect to the max (additive) importance
    Selects all features over the "cutoff" threshold, seen as a multiplicative factor off the normalized max. 
    """
    
    # The list for the additive feature importance vector, initialised to 0.
    add_feats = np.zeros(1167)
    nrModels=0
    
    # Loop through all files in ModelFolder 
    for ind, file in enumerate(os.listdir(ModelFolder)):
        if ".RFmod" not in file:
            continue
        CVmodel = joblib.load(ModelFolder+file)
        add_feats += CVmodel.feature_importances_
        nrModels += 1
        """
        if verbose: 
            print("Model:", file)
            plt.plot(CVmodel.feature_importances_/max(CVmodel.feature_importances_))
            plt.axhline(y=cutoff, color='r', linestyle='-')
            plt.show()
        """

    # Average the additive features per model
    add_feats = add_feats/nrModels
    
    # Normalize it for easier extraction
    add_feats = add_feats/max(add_feats)    
    
    if verbose: 
        print("Top features across the CV models:")
        plt.plot(RAMAN_FREQ_RANGE, ExpandOverSilentRegion(add_feats))
        plt.axhline(y=cutoff, color='r', linestyle='-')
        plt.xlim([RAMAN_FREQ_RANGE[0], RAMAN_FREQ_RANGE[-1]])
        plt.show()
    
    # Take only the features with importances over the threshold
    #add_feats = np.where(add_feats > cutoff, add_feats, 0)
    
    # Get the indices of the important features (score over the cutoff)
    add_feats_sort = np.argsort(add_feats)
    add_feats_importance = add_feats_sort[add_feats[add_feats_sort] > cutoff]
    
    # Undo the silent region removal and sort to observe bands
    add_feats_importance = sorted(UndoSilentRegionRemoval(add_feats_importance))  
    
    # Group the consecutive features into bands/intervals
    add_feats_ranges =[]
    for k,g in groupby(enumerate(add_feats_importance),lambda x:x[0]-x[1]):
        group = (map(itemgetter(1),g))
        group = list(map(int,group))
        add_feats_ranges.append((group[0],group[-1]))
    
    if verbose: 
        print("Top model features:", len(add_feats_importance), "in the ranges:", add_feats_ranges)
    
    return add_feats_importance, add_feats
```
